```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import re
import nltk
from nltk.corpus import stopwords
import warnings
warnings.filterwarnings('ignore')
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
df = pd.read_csv("stock_data.csv", encoding='utf-8')
```

```python
df.head()
```

```python
sentiment_map = {
    -1: 'Negative',
     0: 'Neutral',
     1: 'Positive'
}

df['Sentiment_Label'] = df['Sentiment'].map(sentiment_map)

# Check results
print(df[['Sentiment', 'Sentiment_Label']].head())
print("Missing values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sentiment_Label', data=df, palette='coolwarm')
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
positive_text = " ".join(df[df['Sentiment_Label']=='Positive']['Text'].dropna())
negative_text = " ".join(df[df['Sentiment_Label']=='Negative']['Text'].dropna())
```

```python
wordcloud_pos = WordCloud(width=800, height=400, background_color='white').generate(positive_text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud_pos, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud - Positive Sentiment")
plt.show()
```

```python
wordcloud_neg = WordCloud(width=800, height=400, background_color='white').generate(negative_text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud - Negative Sentiment")
plt.show()
```
